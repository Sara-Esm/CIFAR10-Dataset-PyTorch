```python
# Import necessary libraries
import torch
from torchvision.datasets import CIFAR10
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
import torchvision
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Download and load the CIFAR-10 training dataset
train_data = CIFAR10(root="./train/", train=True, transform=ToTensor(), download=True)
```

```python
# Create a DataLoader to iterate over the dataset
train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
```

```python
# Function to show an image
def show_image(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
# Display the first 16 images
dataiter = iter(train_loader)
images, labels = next(dataiter)  # Use next(iter(...)) to get the next batch
```

```python
# Plot the images
show_image(torchvision.utils.make_grid(images))
```

```python
# Print dataset shape and classes
print(f"Dataset shape: {len(train_data)} samples")
print(f"Classes: {train_data.classes}")
```

```python
# Check class labels
for i in range(16):
    data, label = train_data[i]
    print(f"Picture {i+1}, class: {train_data.classes[label]}")
```
